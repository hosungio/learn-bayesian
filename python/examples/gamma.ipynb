```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import gamma
```

```python
# Note that this parameterization is equivalent to the above, with scale = 1 / beta
params_set = [(1, 1), (2, 1), (3, 1)]
xs = np.linspace(0, 10, 100)
ys_set = [gamma.pdf(x=xs, a=shape, scale=scale) for shape, scale in params_set]
ys_legends = [f"Gamma({shape},{scale})" for shape, scale in params_set]
for ys in ys_set:
    plt.plot(xs, ys)
plt.legend(ys_legends)
plt.show()
```

```python
params_set = [(3, 1), (3, 2), (3, 3)]
xs = np.linspace(0, 10, 100)
ys_set = [gamma.pdf(x=xs, a=shape, scale=scale) for shape, scale in params_set]
ys_legends = [f"Gamma({shape},{scale})" for shape, scale in params_set]
for ys in ys_set:
    plt.plot(xs, ys)
plt.legend(ys_legends)
plt.show()
```

```python
# 1
a = 4
la = 60 / 30 # 단위시간 당 어떤 사건의 평균적인 발생 횟수. 푸아송 분포에서 람다와 동일. 1시간 당 2마리 물고기
scale = 1 / la
gam = gamma(a=a, scale=scale)
print(gam.cdf(4) - gam.cdf(2))
```

```python
# 2
la = 1.6 # 단위시간 당 어떤 사건의 평균적인 발생 횟수. 1분당 평균 1.6개.
a = 20 # 20개의 패널에 대해.
gamma.cdf(x=15, a=a, scale=1/la)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import beta
```

```python

params_set = [(1, 1), (2, 1), (3, 1), (3, 2), (3, 3), (10, 1), (10, 7), (400, 100)]
xs = np.linspace(0, 1, 1000)
ys_set = [beta.pdf(x=xs, a=a, b=b) for a, b in params_set]
ys_legends = [f"B({a},{b})" for a, b in params_set]
for ys in ys_set:
    plt.plot(xs, ys)
plt.legend(ys_legends)
plt.show()
```

```python
params_set = [(1, 1), (1000, 1000)]
xs = np.linspace(0, 1, 1000)
ys_set = [beta.pdf(x=xs, a=a, b=b) for a, b in params_set]
ys_legends = [f"B({a},{b})" for a, b in params_set]
for ys in ys_set:
    plt.plot(xs, ys)
plt.legend(ys_legends)
plt.show()
```

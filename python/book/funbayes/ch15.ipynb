```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
prior_a = 3
prior_b = 7
test_a = beta(36+prior_a, 114+prior_b)
test_b = beta(50+prior_a, 100+prior_b)
n_trials = 100000
a_samples = test_a.rvs(size=n_trials)
b_samples = test_b.rvs(size=n_trials)
print(a_samples.shape, b_samples.shape)
p = sum(b_samples > a_samples) / n_trials
p
```

```python
ba_ratio = b_samples / a_samples
plt.hist(ba_ratio)
plt.show()
```

```python
# https://machinelearningmastery.com/empirical-distribution-function-in-python/
ecdf = ECDF(ba_ratio)
plt.plot(ecdf.x, ecdf.y)
plt.show()
```

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta_3po = beta(a=2, b=7440)
beta_han = beta(a=20000, b=1)
xs = np.linspace(0, 1, 100)
posteriors = beta_3po.pdf(x=xs) + beta_han.pdf(x=xs)
_ = plt.plot(xs, posteriors)
```

```python
beta_joint = beta(a=(2+20000), b=(7440 + 1))
xs = np.linspace(0, 1, 100)
posteriors = beta_joint.pdf(x=xs)
_ = plt.plot(xs, posteriors)
```

```python
# 1
beta1 = beta(a=6, b=1)
prob = beta1.cdf(x=0.6) - beta1.cdf(x=0.4)
print(prob)
```

```python
prior_val = 1
prob = 0
while True:
    beta2 = beta(a=(6+prior_val), b=(1+prior_val))
    prob = beta2.cdf(0.6) - beta2.cdf(0.4)
    if prob >= 0.95:
        break
    else:
        prior_val += 1
print(prior_val, prob)
```

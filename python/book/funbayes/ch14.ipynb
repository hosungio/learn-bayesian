```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from scipy.stats import norm
from scipy.stats import binom
```

```python
prior1 = beta(a=1, b=1)
prior2 = beta(a=1000, b=1000)
xs = np.linspace(0, 1, 100)
ys1 = prior1.pdf(xs)
ys2 = prior2.pdf(xs)
plt.plot(xs, ys1, ".")
plt.plot(xs, ys2)
plt.show()
```

```python
prior1 = beta(7, 3)
prior2 = beta(700, 300)
xs = np.linspace(0, 1, 100)
ys1 = prior1.pdf(xs)
ys2 = prior2.pdf(xs)
plt.plot(xs, ys1, ".")
plt.plot(xs, ys2)
plt.show()
```

```python
# 1
prior1 = beta(7, 3) # 앞면이 나올 확률이 약 70%라고 약하게 믿는 사람
prior2 = beta(1000, 1000) # 앞면이 나올 확률이 50%라고 강하게 믿는 사람
prior3 = beta(700, 300) # 70% 정도 앞면이 나오도록 편향돼 있다고 강하게 믿는 사람
xs = np.linspace(0, 1, 100)
ys1 = prior1.pdf(xs)
ys2 = prior2.pdf(xs)
ys3 = prior3.pdf(xs)
plt.plot(xs, ys1, ".")
plt.plot(xs, ys2, "-")
plt.plot(xs, ys3)
plt.show()
```

```python
# 2
observation_a = 9 + 9
observation_b = 3 + 11
posterior_for_weak_prior = beta(a=observation_a+7, b=observation_b+3)
posterior_for_strong_prior = beta(a=observation_a+70, b=observation_b+30)

xs = np.linspace(0, 1, 100)
weak_ys = posterior_for_weak_prior.pdf(xs)
strong_ys = posterior_for_strong_prior.pdf(xs)
plt.plot(xs, weak_ys)
plt.plot(xs, strong_ys)

weak_lower = posterior_for_weak_prior.ppf(0.025)
weak_upper = posterior_for_weak_prior.ppf(0.975)
strong_lower = posterior_for_strong_prior.ppf(0.025)
strong_upper = posterior_for_strong_prior.ppf(0.975)
print(f"약한 사전에 대한 95% 신뢰구간: {weak_lower} ~ {weak_upper}")
print(f"강한 사전에 대한 95% 신뢰구간: {strong_lower} ~ {strong_upper}")
print(f"약한 사전에 대한 95% 신뢰구간 너비: {weak_upper - weak_lower}")
print(f"강한 사전에 대한 95% 신뢰구간 너비: {strong_upper - strong_lower}")
```

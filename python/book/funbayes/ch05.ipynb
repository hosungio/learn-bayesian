```python
from scipy.stats import binom
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 14
n = 41
ps = np.linspace(0, 1, 100)
probs = [binom.pmf(k=k, n=n, p=p) for p in ps]
_ = plt.plot(ps, probs, ".")
```

```python
ps = np.linspace(0, 1, 100) # 사건의 확률이 얼마일 것이다라는 가설에 대한 각 확률.
a = 14 # 관심있는 사건이 발생한 횟수.
b = 27 # 관심있는 사건이 발생하지 않은 횟수.
y = beta.pdf(x=ps, a=14, b=27)
_ = plt.plot(ps, y, ".")
# 
prob = beta.cdf(x=0.5, a=a, b=b)
print("두 개의 동전이 나올 확률이 0.5 미만일 가능성:", prob)
```

```python
a = 5
b = 1200 - a
beta_dist = beta(a=a, b=b)
xs = np.linspace(0, 0.01, 100)
ys = beta_dist.pdf(x=xs)
_ = plt.plot(xs, ys, ".")
prob = 1 - beta_dist.cdf(x=0.005)
print("브래들리 에프론 카드를 뽑을 가능성이 0.005보다 클 확률:", prob)
```

```python
# 1.
a = 4
b = 6
beta_dist = beta(a=a, b=b)
xs = np.linspace(0, 1, 100)
ys = beta_dist.pdf(x=xs)
_ = plt.plot(xs, ys, ".")
prob = 1 - beta_dist.cdf(x=0.6)
print("동전의 앞면이 60%보다 많을 확률:", prob)
```

```python
# 2.
a = 9
b = 11
beta_dist = beta(a=a, b=b)
xs = np.linspace(0, 1, 100)
ys = beta_dist.pdf(x=xs)
_ = plt.plot(xs, ys, ".")
probs = beta_dist.cdf(x=[0.45, 0.55])
print("5% 이내에서 공정할 확률:", probs[1] - probs[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dices = np.array([4, 6, 8, 12, 20])
priors = np.full(len(dices), 1) / len(dices)
likelihoods = np.array([0, 1/6, 1/8, 1/12, 1/20])
posters = np.multiply(priors, likelihoods)
normalized_posters = posters / posters.sum()
# print(posters)
print(normalized_posters)
```

```python
trains = np.arange(start=1, stop=1001)
priors = np.full(len(trains), 1) / len(trains)
likelihoods = (1 / trains) * np.where(trains < 60, 0, 1)
posters = np.multiply(priors, likelihoods)
normalized_posters = posters / posters.sum()
# print(posters)
print(normalized_posters[55:65])
plt.plot(normalized_posters)
plt.show()
```

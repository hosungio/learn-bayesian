```python
import math
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.arange(start=-5.5, stop=5.6, step=0.1)
ys = [math.exp(-(x**2)/2.0) for x in xs]
plt.plot(xs, ys)
plt.show()
```

```python
xs = np.arange(start=-5.5, stop=5.6, step=0.1)
ys = [math.exp(-(x**2)/2.0)/math.sqrt(math.pi*2) for x in xs]
plt.plot(xs, ys)
plt.show()
```
